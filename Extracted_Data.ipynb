```python
import os
import json
import pandas as pd
```

```python
#To get the aggregation Transaction data from the overall data
direct1 = r"C:\Users\91809\pulse\data\aggregated\transaction\country\india\state"
data = []

#Create the loop for to merge the all path and directories
for state_dir in os.listdir(direct1):
    state_path = os.path.join(direct1,state_dir)
    for year_dir in os.listdir(state_path):
        year_path = os.path.join(state_path,year_dir)
        for quart_file in os.listdir(year_path):
            quart_path = os.path.join(year_path,quart_file)
            with open(quart_path,'r') as f:
                quart_data = json.load(f)
            for trans in quart_data['data']['transactionData']:
                name= trans['name']
                count = trans['paymentInstruments'][0]['count']
                amount=trans['paymentInstruments'][0]['amount']
                data.append({
                    'State':state_dir,
                    'Year':year_dir,
                    'Quarter':int(os.path.splitext(quart_file)[0]),
                    'Transaction_type':name,
                    'Transaction_count':count,
                    'Transaction_amount':amount
                })
#Create the datafarme for 
agg_transaction = pd.DataFrame(data)
agg_transaction.head(5)
```

```python
agg_transaction.to_csv('Agg_trans.csv',index=False)
```

```python
#To get the aggregation user data from the overall data
direct2=r"C:\Users\91809\pulse\data\aggregated\user\country\india\state"

# Create an empty list to store the data 
data = []
#create the loop for merge the directories from various paths

for state_dir in os.listdir(direct2):
    state_path = os.path.join(direct2,state_dir)
    for year_dir in os.listdir(state_path):
        year_path = os.path.join(state_path,year_dir)
        for quart_file in os.listdir(year_path):
            quart_path = os.path.join(year_path,quart_file)
            with open(quart_path,'r') as f:
                quart_data = json.load(f)
            try:
                for user in quart_data['data']['usersByDevice']:
                    brand = user['brand']
                    count = user['count']
                    percentage = user['percentage']
                    data.append({
                        'State':state_dir,
                        'Year':year_dir,
                        'Quarter':int(os.path.splitext(quart_file)[0]),
                        'Brand':brand,
                        'Brand_count':count,
                        'Brand_percentage':percentage
                    })
            except:
                pass
#Create dataframe 
agg_user = pd.DataFrame(data)
agg_user.head(5)
                
```

```python
agg_user.to_csv('Agg_user.csv',index=False)
```
